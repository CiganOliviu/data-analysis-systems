```python
import pandas as pd
```

```python
data_frame = pd.read_csv ("Coastal_Data_System.csv", index_col = 0)
```

```python
data_frame.dropna()
```

```python
from sklearn import svm, preprocessing
import sklearn
```

```python
data_frame = sklearn.utils.shuffle(data_frame)
```

```python
Xs = data_frame.drop("Hs", axis = 1).values
```

```python
Xs = preprocessing.scale(Xs)
```

```python
Ys = data_frame["Hs"].values
```

```python
Ys
```

```python
test_size = 100
```

```python
Xs_train = Xs[:-test_size]
Ys_train = Ys[:-test_size]
Xs_test = Xs[-test_size:]
Ys_test = Ys[-test_size:]
```

```python
classifier = svm.SVR(kernel="rbf", gamma = "scale")
```

```python
classifier.fit (Xs_train, Ys_train)
```

```python
classifier.score(Xs_test, Ys_test)
```

```python
for Xs, Ys in zip(Xs_test, Ys_test):
    print(f"Model {classifier.predict([Xs])[0]}, Actual {Ys}")
```
